### 1. How many are there in each team and the percentage splitting with respect to the total employees.

```python
team_count=df['Team'].value_counts()
team_count
```

```python
team_len=len(df)
team_len
```

```python
team_percentage=(team_count/team_len)*100
team_percentage
```

### 2. Segregate the employees with respect to different positions.

```python
x=df.groupby('Position')
for position, position_data in x:
    print(f'Position:{position}')
    print(position_data[['Name','Team',]])
```

### 3. Find out under which age group most of the employees belong to.

```python
age_count=df['Age'].value_counts()
age_count
```

```python
print(f"Most of the employees belongs to the group:{age_count.idxmax()}")
print(f"No.of employees:{age_count.max()}")
```

### 4. Find out under which team & position spending in terms of salary is high.

```python
highest_salary_team=df.groupby(['Team','Position'])['Salary'].sum()
team_position=highest_salary_team.idxmax()
salary=highest_salary_team.max()
print(f'Team spending highest salary:{team_position}')
print(f'Highest salary:{salary}')
```

### 5. Find if there is any correlation between age & salary, represent it visually.

```python
correlation=df['Age'].corr(df['Salary'])
print(f'Correlation coefficient between age & salary:{correlation}')
```

### Graphical representation of correlation between age & salary.

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,8))
plt.scatter(df['Age'],df['Salary'])
plt.title('Scatter Plot of Age & Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```
